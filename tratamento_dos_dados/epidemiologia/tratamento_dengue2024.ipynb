```python
df['DATA DE NASCIMENTO'] = pd.to_datetime(df['DATA DE NASCIMENTO'], format='%m/%d/%y', errors='coerce')

# Correcting years that are in the future (e.g., '25' interpreted as 2025 instead of 1925)
def adjust_year(date_series):
    current_year = pd.Timestamp.now().year
    adjusted_dates = []
    for date in date_series:
        if pd.isna(date):
            adjusted_dates.append(pd.NaT)
            continue
        if date.year > current_year:
            adjusted_dates.append(date - pd.DateOffset(years=100))
        else:
            adjusted_dates.append(date)
    return pd.Series(adjusted_dates, index=date_series.index)

df['DATA DE NASCIMENTO'] = adjust_year(df['DATA DE NASCIMENTO'])

# Calculate age in years
df['IDADE EM ANOS'] = (df['DATA REGISTRO'] - df['DATA DE NASCIMENTO']).dt.days / 365.25

# Round the age to the nearest integer
df['IDADE EM ANOS'] = df['IDADE EM ANOS'].round().astype('Int64')

display(df[['DATA DE NASCIMENTO', 'DATA REGISTRO', 'IDADE EM ANOS']].head())
```

# Task
Calculate the proportion of 'DENV 1' and 'DENV 2' in the 'SOROTIPO' column, excluding NaN values, then use these proportions to randomly fill the NaN values in the same column. Finally, verify the updated counts to confirm the distribution.

## Calculate Sorotype Proportions

### Subtask:
Calculate the current proportions of 'DENV 1' and 'DENV 2' in the 'SOROTIPO' column, excluding 'nan' values.


**Reasoning**:
To calculate the proportions of 'DENV 1' and 'DENV 2', I need to get the value counts of the 'SOROTIPO' column, exclude 'nan' entries, and then divide each sorotype's count by the total non-nan count.



```python
sorotipo_counts = df_final2024['SOROTIPO'].value_counts(dropna=True)
display(sorotipo_counts)

total_sorotipos = sorotipo_counts.sum()

denv1_proportion = sorotipo_counts.get('DENV 1', 0) / total_sorotipos if total_sorotipos > 0 else 0
denv2_proportion = sorotipo_counts.get('DENV 2', 0) / total_sorotipos if total_sorotipos > 0 else 0

print(f"Proportion of DENV 1: {denv1_proportion:.4f}")
print(f"Proportion of DENV 2: {denv2_proportion:.4f}")
```

**Reasoning**:
Now that the proportions of 'DENV 1' and 'DENV 2' have been calculated, the next step is to randomly fill the NaN values in the 'SOROTIPO' column based on these proportions. This will maintain the observed distribution.



```python
nan_indices = df_final2024['SOROTIPO'].isna()
num_nan = nan_indices.sum()

if num_nan > 0:
    # Create a list of possible sorotypes based on their proportions
    choices = ['DENV 1', 'DENV 2']
    probabilities = [denv1_proportion, denv2_proportion]

    # Randomly select sorotypes for the NaN values based on the calculated proportions
    random_sorotypes = np.random.choice(choices, size=num_nan, p=probabilities)

    # Fill the NaN values in the 'SOROTIPO' column
    df_final2024.loc[nan_indices, 'SOROTIPO'] = random_sorotypes

# Verify the updated counts and proportions
updated_sorotipo_counts = df_final2024['SOROTIPO'].value_counts(dropna=False)
display(updated_sorotipo_counts)

updated_total_sorotipos = updated_sorotipo_counts.sum()
updated_denv1_proportion = updated_sorotipo_counts.get('DENV 1', 0) / updated_total_sorotipos
updated_denv2_proportion = updated_sorotipo_counts.get('DENV 2', 0) / updated_total_sorotipos

print(f"Updated Proportion of DENV 1: {updated_denv1_proportion:.4f}")
print(f"Updated Proportion of DENV 2: {updated_denv2_proportion:.4f}")
```
